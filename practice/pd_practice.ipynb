```python
my_stat['V5'] = my_stat.V1 + my_stat.V4
my_stat['V6'] = np.log(my_stat['V2'])
my_stat
```

```python
my_stat = my_stat.rename(columns={'V1': 'session_value', 'V2': 'group', 'V3': 'time', 'V4': 'n_users'})
my_stat
```

```python
my_stat_1 = pd.read_csv('C:/Users/User/Desktop/dtml/my_stat_1.csv')
my_stat_1
```

```python
my_stat_1 = my_stat_1.fillna(0)
my_stat_2 = my_stat_1.query('n_users >= 0')
med = my_stat_2['n_users'].median()
my_stat_1.loc[my_stat_1['n_users']<0, 'n_users'] = med 
my_stat_1
```

```python
mean_session_value_data = my_stat_1.groupby('group', as_index=False).agg({'session_value': 'mean'})
mean_session_value_data.rename(columns={'session_value': 'mean_session_value'}, inplace=True)
mean_session_value_data
```
