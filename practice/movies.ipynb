```python
import numpy as np
import pandas as pd
import timeit
```

```python
movie = pd.read_csv('C:/Users/User/Desktop/dtml/movie_metadata.csv')
movie
```

```python
genres = movie[['director_name', 'genres']]
```

```python
genres
```

```python
for i in genres:
    print(i)
```

```python
for i in genres.values:
    print(i)
```

```python
for row in genres.values:
    for value in row:
        print(value)
```

```python
def reversator(value):
    return value[::-1]
```

```python
for row in genres.values:
    for value in row:
        print(reversator(value))
```

```python
for row in genres.iterrows():
    print(row)
```

```python
for i,row in genres.iterrows():
    print(row)
```

```python
for i,row in genres.iterrows():
    print(row.map(reversator))
```

```python
for i,col in genres.iteritems():
    print(col)
```

```python
for col in genres.itertuples():
    print(reversator(col))
```
